```python
#Importing important libraries

import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from collections import Counter
from sklearn.preprocessing import Normalizer
```

```python
irisData=load_iris()
iris= pd.DataFrame(irisData.data)
targetColumns=pd.DataFrame(irisData.target)
```

```python
iris.head()
```

```python
X = irisData.data
y = irisData.target
```

```python
#Spilt data into test & train
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,shuffle=True, random_state=42)
```

```python
#Apply KNN Algorithm
KNN = KNeighborsClassifier(n_neighbors= 10)
KNN.fit(X_train,y_train)
```

```python
#Prediction Using KNN
y_pred= KNN.predict(X_test)
y_pred
```

```python
#Finding Accuracy
accuracyOfModel = accuracy_score(y_test,y_pred)
accuracyOfModel
```

** Raw Implementation Without Normalization **

```python
#Finding distance between test & trained data

def dataDistance(trainedPoint , testPoint):
  distanceOfPoint = []
  for row in range(len(trainedPoint)):
      currentTrainPoint= trainedPoint[row]
      currentDistance= 0
      for col in range(len(currentTrainPoint)):         
          currentDistance += (currentTrainPoint[col] - testPoint[col]) **2
      currentDistance= np.sqrt(currentDistance)
      distanceOfPoint.append(currentDistance)
  distanceOfPoint= pd.DataFrame(data=distanceOfPoint,columns=['distanceCols'])
  return distanceOfPoint
```

```python
def minimalDistanceOfTestAndTrain(distanceOfPoint, totalNeighborsK):
    nearestNeighbors= distanceOfPoint.sort_values(by=['distanceCols'], axis=0)
    nearestNeighbors= nearestNeighbors[:totalNeighborsK]
    return nearestNeighbors
```

```python
def mostDataPoint(nearestNeighbors, y_train):   
    majorPoints= Counter(y_train[nearestNeighbors.index])
    y_pred= majorPoints.most_common()[0][0]
    return y_pred
```

```python
def rawKNN(x_train, y_train, x_test, K):
    y_pred=[]
    for eachTestPoint in x_test:
      distancePoint  = dataDistance(x_train, eachTestPoint)
      nearestPoint= minimalDistanceOfTestAndTrain(distancePoint, K)
      predictionPoint = mostDataPoint(nearestPoint, y_train)
      y_pred.append(predictionPoint)

    return y_pred  
```

```python
K = 10
predShihab = rawKNN(X_train, y_train, X_test, K)
predShihab
```

```python
accuracyOfmyModel = accuracy_score(y_test,predShihab)
accuracyOfmyModel
```
